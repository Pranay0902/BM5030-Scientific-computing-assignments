## Task_1:Part 1
### For this we will plot population of 3 countries aganist years and see which kind of graph is best for seeing the trend over years and comparing between themself.

```python
# First import libraries
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = [2020,2019,2018,2017,2016,2015,2014,2013,2012,2011,2010]
# Population of Brazil
y1 = [212559417,211049527,209469323,207833823,206163053,204471769,202763739,201035912,199287299,197514536,195713635]
# plotting the line 1 points 
plt.plot(x, y1, label = "Brazil")
# Population of Russia
y2 = [145934462,145872256,145734038,145530082,145275383,144985057,144664841,144325449,143993892,143703025,143479274]
# plotting the line 2 points 
plt.plot(x, y2, label = "Russia")
# Population of USA
y3 = [331002651,329064917,327096265,325084756,323015995,320878310,318673411,316400538,314043885,311584047,309011475]
# plotting the line 3 points 
plt.plot(x, y3, label = "USA")
plt.xlabel('Year')
# Set the y axis label of the current axis.
plt.ylabel('Population(x10^8)')
# Set a title of the current axes.
plt.title('Population of countries over years')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

### The line graph does show us how the trends of population of countries are going over the years. So it is a good way but not the best because we don't know how the trends are when we include all countries 

```python
plt.scatter(x, y1, label = "Brazil")
plt.scatter(x, y2, label = "Russia")
plt.scatter(x, y3, label = "USA")
plt.xlabel('Year')
# Set the y axis label of the current axis.
plt.ylabel('Population(x10^8)')
# Set a title of the current axes.
plt.title('Population of countries over years')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

### Comparing everything the scatterplot is the worst possible representaion as it is almost similar to line in this case but is even less informative than line graph

```python
N = 11
  
ind = np.arange(N)    
width = 0.35  
  
fig = plt.subplots(figsize =(10, 7)) 
p1 = plt.bar(ind, y1, width) 
p2 = plt.bar(ind, y2, width, bottom = y1) 
p3 = plt.bar(ind, y3, width, bottom = y3)
plt.xlabel('Years')  
plt.ylabel('Population(x10^9)') 
plt.title('Population of countries over years') 
plt.xticks(ind, (2020,2019,2018,2017,2016,2015,2014,2013,2012,2011,2010)) 
plt.yticks(np.arange(0, 81, 10)) 
plt.legend((p1[0], p2[0], p3[0]), ('Brazil', 'Russia','USA')) 
  
plt.show()
```

### This representation is preety good because it shows us the overall trend and it also is good in comparing between any 2 countries

```python
fig = plt.subplots(figsize =(10, 7)) 
X = np.arange(11)
plt.bar(X - 0.25, y1, color = 'b', width = 0.25)
plt.bar(X + 0.00, y2, color = 'g', width = 0.25)
plt.bar(X + 0.25, y3, color = 'r', width = 0.25)
plt.xlabel('Years')  
plt.ylabel('Population(x10^9)') 
plt.xticks(ind, (2020,2019,2018,2017,2016,2015,2014,2013,2012,2011,2010)) 
plt.legend(('Brazil', 'Russia','USA'))
plt.show()
```

### Finally, out of all this is the best representation because shows the indivisual trends of each country compares them and also roughly gives the trend of overall population

## Task_2 : Part 2
### But if we want to see the relation between temparature and humidity we need to know the count for a given temparature and humidity and we will use a 3d histogram for that

```python
xAmplitudes = data_2d[0]
yAmplitudes = data_2d[1]

x = np.array(data_2d[0])   #turn x,y data into numpy arrays
y = np.array(data_2d[1])

fig = plt.figure(figsize =(10, 7))          #create a canvas, tell matplotlib it's 3d
ax = fig.add_subplot(111, projection='3d')

#make histogram stuff - set bins - I choose 20x20 because I have a lot of data
hist, xedges, yedges = np.histogram2d(x, y, bins=(20,20))
xpos, ypos = np.meshgrid(xedges[1:],yedges[1:])

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like (xpos)

dx = xedges [1] - xedges [0]
dy = yedges [1] - yedges [0]
dz = hist.flatten()

#cmap = cm.get_cmap('jet') # Get desired colormap - you can change this!
max_height = np.max(dz)   # get range of colorbars so we can normalize
min_height = np.min(dz)
# scale each z to [0,1], and get their rgb values
#rgba = [cmap((k-min_height)/max_height) for k in dz] 

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
plt.xlabel("Temp")
plt.ylabel("Humidity")
plt.savefig("Relation between temp and humidity")
plt.show()
```

### From this we can see very clearly that for low temparature most of the days have low humidity and for high temparature most of the days have high humidity. Looking at the graph we can see that the peaks are along the diagonal which represents that Temparature is directly proportional to Humidity
